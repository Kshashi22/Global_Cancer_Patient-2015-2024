```python
ax = sns.countplot(x='Gender',data=data)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='Country_Region',data=data)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='Cancer_Type',data=data, palette='rocket')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
print(data.columns)
Cancer_Type= data['Cancer_Type'].value_counts(normalize=True)
print(Cancer_Type)

plt.figure(figsize=(5,4))
plt.pie(Cancer_Type.values,labels=Cancer_Type.index,autopct='%1.1f%%',startangle=90)
plt.title('Cancer_Type')
plt.show()
```

```python
print(data.columns)
Cancer_Stage= data['Cancer_Stage'].value_counts(normalize=True)
print(Cancer_Stage)

plt.figure(figsize=(5,4))
plt.pie(Cancer_Stage.values,labels=Cancer_Stage.index,autopct='%1.1f%%',startangle=90)
plt.title('Cancer_Stage')
plt.show()
```
